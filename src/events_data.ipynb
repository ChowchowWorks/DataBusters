```python
import pandas as pd
import numpy as np 
from datetime import datetime
```

```python
tk1 = pd.read_csv('../data/event_data.csv', encoding='latin1')
```

```python
tk1['datetime'] = pd.to_datetime(tk1['timestamp'], unit='s')
# Extract date components
tk1['date'] = tk1['datetime'].dt.date
tk1['year'] = tk1['datetime'].dt.year
tk1['month'] = tk1['datetime'].dt.month
tk1['month_name'] = tk1['datetime'].dt.month_name()
tk1['day'] = tk1['datetime'].dt.day
tk1['hour'] = tk1['datetime'].dt.hour
tk1['minute'] = tk1['datetime'].dt.minute

# View results
print(tk1[['timestamp', 'datetime', 'date', 'month', 'month_name']].head())
print(tk1[['timestamp', 'datetime', 'date', 'month', 'month_name']].tail())
```

```python
tk1.drop(columns = ['date','year','month','month_name','day','hour','minute'])
```
